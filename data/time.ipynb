```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/time_taking.csv")
```

```python
data.head()
```

```python

data_counter = data.groupby("counter").mean()
avg_data = data_counter.drop(["M","total_time"], axis=1)
data_counter.head()
```

```python
#Som vi ser her s brukes meste parten av tiden p rolloutsene men de synker i verdi siden slutt handlingen intreffer tidligere
plt.plot(avg_data)
plt.legend(["tree search", "rollout"])
plt.xlabel("actions taken")
plt.ylabel("time s")
plt.show()
```

```python
check_values = data_counter
check_values["treesearch"] = check_values["treesearch"]*check_values["M"]
check_values["rollout"] = check_values["rollout"]*check_values["M"]
```

```python
plt.plot(check_values.drop("M",axis=1))
plt.legend(["total time","tree search", "rollout"])
plt.xlabel("actions taken")
plt.ylabel("time s")
plt.show()
```
